```python
import pandas as pd

df = pd.read_csv("07561.csv")
df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


df["Re(Z)"] = df["Battery_impedance"].apply(lambda x: complex(x).real)
df["Im(Z)"] = df["Battery_impedance"].apply(lambda x: complex(x).imag)
df["Cycle"] = np.arange(1, len(df) + 1)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


ax.plot(df["Re(Z)"], df["Im(Z)"], df["Cycle"], marker='o', linestyle='-', lw=1, label="EIS Data")
ax.set_xlabel("Re(Z) [Ohm]")
ax.set_ylabel("Im(Z) [Ohm]")
ax.set_zlabel("Cycle Count")
ax.set_title("3D Nyquist Plot of EIS Data")
ax.legend()
plt.show()
```
